# 課題1
以下の仕様に従って、長方形を表す Rectangle クラスを定義してください。



*   コンストラクタで width（幅）と height（高さ）を受け取る。
*   面積を返す area() メソッドを持つ。
*   周囲の長さを返す perimeter() メソッドを持つ。


```python
class Rectangle:
  def __init__(self, width, height):
    self.width = width
    self.height = height

  def area(self):
    return self.width * self.height

  def perimeter(self):
    return 2 * (self.width + self.height)
```

# 課題2

銀行口座を表す BankAccount クラスを作成してください。

*  コンストラクタで owner（口座名義）と初期残高 balance を受け取る。
*  deposit(amount) メソッドで入金。
*  withdraw(amount) メソッドで出金（残高不足なら何もしない）。
*  get_balance() メソッドで残高を返す。

```python
class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance
  def deposit(self, amount):
    self.balance += amount
  def withdraw(self, amount):
    if self.balance >= amount:
      self.balance -= amount
  def get_balance(self):
    return self.balance
```

# 課題3
学生の成績を管理する Student クラスを作成してください。

*  コンストラクタで name（名前）を受け取る。
*  add_score(subject, score) メソッドで科目と得点を追加。
*  average_score() メソッドで平均点を返す。
*  get_score(subject) メソッドで特定科目の得点を返す。

```python
class Student:
  def __init__(self, name):
    self.name = name
    self.scores = {}
  def add_score(self, subject, score):
    self.scores[subject] = score
  def average_score(self):
    return sum(self.scores.values()) / len(self.scores)
  def get_score(self, subject):
    return self.scores.get(subject)
```

# 課題4
多項式を表す Polynomial クラスを作成してください。

*  コンストラクタで係数リスト（例：[1, 0, -2] は $ x^2-2 $ を受け取る。
*  evaluate(x) メソッドで、指定した値での多項式の値を返す。
*  __str__() メソッドで人間が読める形で多項式を表示（例: x^2 - 2）。

```python
class Polynomial:
  def __init__(self, lst):
    self.lst = lst
  def evaluate(self, x):
    res = ""
    N = len(self.lst)
    for i in range(len(self.lst)):
      if i == N-1:
        if self.lst[i] > 0:
          res += "+" + str(self.lst[i])
          continue
        elif self.lst[i] == 0:
          continue
        else:
          res += str(self.lst[i])
      elif i == N-2:
        if self.lst[i] == -1:
          res += "-" + "x^" + str(N-1-i)
          continue
        if self.lst[i] == 0:
          continue
        if self.lst[i] == 1:
          res += str(x) + "^" + str(N-1-i)
          continue
        if self.lst[i] > 0:
          res += str(self.lst[i]) + str(x) + "^" + str(N-1-i)
          continue
      elif i == 0:
        if self.lst[i] == -1:
          res += "-" + "x^" + str(N-1-i)
          continue
        if self.lst[i] == 0:
          continue
        if self.lst[i] == 1:
          res += str(x) + "^" + str(N-1-i)
          continue
        else:
          res += str(self.lst[i]) + str(x) + "^" + str(N-1-i)

      else:
        if self.lst[i] == 0:
          continue
        if self.lst[i] == 1:
          res += "+" + str(self.lst[i]) + "x^" + str(N-1-i)
          continue
        if self.lst[i] == -1:
          res += "-" + "x^" + str(N-1-i)
          continue
        else:
          res += str(self.lst[i]) + str(x) + "^" + str(N-1-i)
          continue
    return res

c = Polynomial([-1, -1, 1])
print(c.evaluate("x"))


```
